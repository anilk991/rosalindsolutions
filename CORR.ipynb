```python
s1 = 'TTCAT'
s2 = 'TTGAT'
```

```python
def compare(s1, s2):
    '''
    '''
    misCount = 0
    for i in range(len(s1)):
        if s1[i] != s2[i]:
            misCount += 1
    if misCount > 1:
        return False
    elif misCount == 0:
        return 1
    elif misCount == 1:
        return True
```

```python
def rev_comp(dna):
    '''
    '''
    comps = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    return ''.join(comps[base] for base in dna[::-1])
```

```python
reads = ['TCATC', 'TTCAT', 'TCATC', 'TGAAA', 'GAGGA', 'TTTCA', 'ATCAA', 'TTGAT', 'TTTCC']
```

```python
def read_counts(reads):
    '''
    '''
    readC = {}
    for i in range(len(reads)):
        count = 1
        for j in range(len(reads)):
            if i != j:
                if reads[i] == reads[j] or reads[i] == rev_comp(reads[j]):
                    count += 1
        if reads[i] not in readC.keys():
            readC[reads[i]] = count
            
    return readC
```

```python
read_counts(reads)
```

```python
def mis_cor_reads(reads):
    '''
    '''
    mis_reads = []
    cor_reads = []
    for read in readC.keys():
        if readC[read] < 2:
            mis_reads.append(read)
        elif readC[read] >= 1:
            cor_reads.append(read)
            
    return cor_reads, mis_reads
```

```python
def correct(reads):
    '''
    '''
    
    readC = read_counts(reads)
    
    cor_reads, mis_reads = mis_cor_reads(reads)
            
    for m in mis_reads:
        #find = False
        for c in cor_reads:
            rev = rev_comp(c)
            if compare(m, c) and not compare(m, rev):
                #find = True
                print(m+"->"+c)
                break
            elif not compare(m, c) and compare(m, rev):
                #find = True
                print(m+"->"+rev)
                break
```

```python
correct(reads)
```

```python
def from_file(path):
    '''
    '''
    handle = open(path)
    reads = []
    for record in SeqIO.parse(handle, "fasta"):
        reads.append(str(record.seq))
    
    return reads
```
