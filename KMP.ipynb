```python
s = 'CAGCATGGTATCACAGCAGAG'
```

```python
for i,c in enumerate(s):
    print(i, c)
```

```python
F_array = [0] * len(s)
k = 0
for i in range(2, len(s) + 1):
    while k > 0 and s[k] != s[i - 1]:
        k = F_array[k - 1]
        #print(*F_array)
    if s[k] == s[i - 1]:
        k += 1
        #print(*F_array)
    F_array[i - 1] = k
```

```python
print(*F_array)
```
