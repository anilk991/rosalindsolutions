```python
contigs = ['ATTAGACCTG', 'CCTGCCGGAA', 'AGACCTGCCG', 'GCCGGAATAC']
```

```python
def overlap(reads, genome=''):
    '''
    '''
    if len(reads) == 0:
        return genome
    
    elif len(genome) == 0:
        genome = reads.pop(0)
        return overlap(reads, genome)
    
    else:
        for i in range(len(reads)):
            read = reads[i]
            l = len(read)
            
            for p in range(int(l/2)):
                q = l - p
                
                if genome.startswith(read[p:]):
                    reads.pop(i)
                    return overlap(reads, read[:p] + genome)
                
                if genome.endswith(read[:q]):
                    reads.pop(i)
                    return overlap(reads, genome + read[q:])
```

```python
overlap(contigs)
```

```python
handle = open("/Users/anilkumar/projectEEG/rosalind_long.txt")
reads = []
for record in SeqIO.parse(handle, "fasta"):
    reads.append(str(record.seq))
handle.close()
```
