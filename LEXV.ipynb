```python
import itertools
```

```python
s = 'DNA'
n = 3
```

```python
def lexv(alphabets, n):
    '''
    '''
    perms = []
    for i in range(1, n + 1):
        for j in itertools.product(alphabets, repeat=i):
            perms.append(''.join(j))
    
    srt_perms = sorted(perms, key = lambda word: [alphabets.index(c) for c in word])
    
    return srt_perms
```

```python
for i in lexv(s, 3):
    print(i)
```

```python
exA = 'KAXGWYBHVCJP'
n = 3
```

```python
for i in lexv(exA, n):
    print(i)
```
