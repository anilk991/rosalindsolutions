```python
reads = ['TGAT', 'CATG', 'TCAT','ATGC','CATC','CATC']
```

```python
def rev_comp(dna):
    '''
    '''
    comps = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A'}
    return "".join(comps[base] for base in dna[::-1])
```

```python
rc_reads = [rev_comp(read) for read in reads]
```

```python
tot_reads = set(reads + rc_reads)
```

```python
def deBruijin(reads):
    '''
    '''
    rc_reads = [rev_comp(read) for read in reads]
    tot_reads = set(reads +  rc_reads)
    
    for read in tot_reads:
        print('('+read[:-1]+', '+ read[1:]+')')
```

```python
deBruijin(reads)
```

```python
def read_file(path):
    '''
    '''
    h = open(path).readlines()
    h = [line.strip() for line in h]
    
    return h
```
